### Create more descriptive/accurate TYPE values according to the bussiness name

```python
def classify_final_type(name, current_type):
    name = str(name).lower().strip()
    
    # Step 1: Override for clubs
    if current_type == "Club":
        return "Social Club"
    
    # Step 2: Handle bars or lounges
    elif current_type == "General On-Premises":
        return "Bar or Lounge"
    
    # Step 3: Restaurants
    elif current_type == "Restaurant":
        return "Restaurant"
    
    # Step 4: Convenience store stays as is
    elif current_type == "Convenience Store":
        return "Convenience Store"
    
    # Step 5: Refine Package Stores
    elif current_type == "Package Store":
        # Grocery store patterns
        if any(keyword in name for keyword in ["market", "supermarket", "meat", "grocery"]):
            return "Grocery Store w/ Liquor"
        # Convenience-like patterns misclassified as package
        elif any(keyword in name for keyword in ["mini", "mart", "corner", "stop", "store", "chicken"]):
            return "Convenience Store"
        # True liquor stores
        elif any(keyword in name for keyword in ["liquor", "spirits", "wine", "package"]):
            return "Liquor Store"
        else:
            return "Liquor Store"
    
    # Step 6: Fallback
    else:
        return "Other"

# Apply to your DataFrame
filtered_liquor_df["TYPE"] = filtered_liquor_df.apply(
    lambda row: classify_final_type(row["NAME"], row["TYPE"]),
    axis=1
)
```

### Geocode the FULL_PREMISE_ADDRESS column

```python
import pandas as pd
import os
import time
import json
from opencage.geocoder import OpenCageGeocode

# -------------------------
# 🔑 Load OpenCage API Key
# -------------------------
notebook_dir = os.getcwd()
config_path = os.path.join(notebook_dir, '..', 'config.json')  # adjust if needed
with open(config_path, 'r') as f:
    config = json.load(f)

key = config['OPENCAGE_API_KEY']
geocoder = OpenCageGeocode(key)

# -------------------------
# 📂 Load Data
# -------------------------
# If you already have filtered_liquor_df defined, skip this
# filtered_liquor_df = pd.read_csv("your_file.csv")

# Add empty latitude and longitude columns if not already there
if 'latitude' not in filtered_liquor_df.columns:
    filtered_liquor_df['latitude'] = None
if 'longitude' not in filtered_liquor_df.columns:
    filtered_liquor_df['longitude'] = None

# -------------------------
# 📍 Geocode Missing Coordinates
# -------------------------
missing = filtered_liquor_df[filtered_liquor_df['latitude'].isna() | filtered_liquor_df['longitude'].isna()]

for idx, row in missing.iterrows():
    address = row['FULL_PREMISE_ADDRESS']
    
    try:
        response = geocoder.geocode(address)
        if response and len(response) > 0:
            lat = response[0]['geometry']['lat']
            lon = response[0]['geometry']['lng']
            filtered_liquor_df.at[idx, 'latitude'] = lat
            filtered_liquor_df.at[idx, 'longitude'] = lon
            print(f"Geocoded: {address} -> ({lat}, {lon})")
        else:
            print(f"No result for: {address}")
    except Exception as e:
        print(f"Error for {address}: {e}")
    
    time.sleep(1.2)  # respects API rate limit

# -------------------------
# 💾 Save to CSV (optional)
# -------------------------
output_path = os.path.join(notebook_dir, '..', 'data', 'secondary_data', 'filtered_liquor_geocoded.csv')
filtered_liquor_df.to_csv(output_path, index=False)
print("✅ Geocoding complete. File saved.")
```

### Create a map of the liquor retail locations, output to HTML

```python
import pandas as pd
import os

notebook_dir = os.getcwd()
current_liquor_df = pd.read_csv(os.path.join(notebook_dir, '..', 'data', 'secondary_data', 'liquor_retail_geocoded.csv'))
```

```python
import folium

# Drop rows with missing coordinates
filtered_liquor_df = current_liquor_df.dropna(subset=["latitude", "longitude"])

# ---------------------------
# 3. Filter by TYPE
# ---------------------------

restaurant_df = filtered_liquor_df[filtered_liquor_df["TYPE"] == "Restaurant"]
liquor_df = filtered_liquor_df[filtered_liquor_df["TYPE"] == "Liquor Store"]
bar_df = filtered_liquor_df[filtered_liquor_df["TYPE"] == "Bar or Lounge"]
grocery_df = filtered_liquor_df[filtered_liquor_df["TYPE"] == "Grocery Store w/ Liquor"]
convenience_df = filtered_liquor_df[filtered_liquor_df["TYPE"] == "Convenience Store"]
social_df = filtered_liquor_df[filtered_liquor_df["TYPE"] == "Social Club"]
nightclub_df = filtered_liquor_df[filtered_liquor_df["TYPE"] == "Nightclub"]

# ---------------------------
# 4. Create the Map
# ---------------------------

m = folium.Map(location=[42.707, -71.163], zoom_start=13)

# Add markers by type
for _, row in restaurant_df.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=f'{row["NAME"]} (Restaurant)',
        tooltip=row["NAME"],
        icon=folium.Icon(color="red", icon="cutlery")
    ).add_to(m)

for _, row in liquor_df.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=f'{row["NAME"]} (Liquor Store)',
        tooltip=row["NAME"],
        icon=folium.Icon(color="darkgreen", icon="shopping-cart")
    ).add_to(m)

for _, row in bar_df.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=f'{row["NAME"]} (Bar or Lounge)',
        tooltip=row["NAME"],
        icon=folium.Icon(color="blue", icon="glass")
    ).add_to(m)

for _, row in nightclub_df.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=f'{row["NAME"]} (Nightclub)',
        tooltip=row["NAME"],
        icon=folium.Icon(color="black", icon="music")
    ).add_to(m)

for _, row in grocery_df.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=f'{row["NAME"]} (Grocery Store w/ Liquor)',
        tooltip=row["NAME"],
        icon=folium.Icon(color="green", icon="shopping-cart")
    ).add_to(m)

for _, row in convenience_df.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=f'{row["NAME"]} (Convenience Store)',
        tooltip=row["NAME"],
        icon=folium.Icon(color="purple", icon="shopping-cart")
    ).add_to(m)

for _, row in social_df.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=f'{row["NAME"]} (Social Club)',
        tooltip=row["NAME"],
        icon=folium.Icon(color="orange", icon="star")
    ).add_to(m)


# ---------------------------
# 5. Add Legend
# ---------------------------

legend_html = '''
<div style="
    position: fixed;
    bottom: 40px;
    left: 40px;
    width: 280px;
    background-color: white;
    border:2px solid grey;
    z-index:9999;
    font-size:14px;
    padding: 10px;
    box-shadow: 2px 2px 5px rgba(0,0,0,0.3);
">
<b>Legend</b><br>
<i style="color:red;">⬤</i> Restaurant<br>
<i style="color:darkgreen;">⬤</i> Liquor Store<br>
<i style="color:blue;">⬤</i> Bar or Lounge<br>
<i style="color:black;">⬤</i> Nightclub<br>
<i style="color:green;">⬤</i> Grocery Store w/ Liquor<br>
<i style="color:purple;">⬤</i> Convenience Store<br>
<i style="color:orange;">⬤</i> Social Club
</div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# ---------------------------
# 6. Save the Map
# ---------------------------

m.save("current_liquor_license_map_by_type.html")
```
